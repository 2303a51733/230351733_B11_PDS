```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
countries = ['China', 'India', 'USA', 'Indonesia', 'Pakistan']
populations = [1412000000, 1393000000, 332000000, 276000000, 225000000]
plt.figure(figsize=(10, 6))
bars = plt.bar(countries, populations, color='skyblue')
def format_population(x, pos):
    if x >= 1e9:
        return f'{x * 1e-9:.1f}B'
    elif x >= 1e6:
        return f'{x * 1e-6:.0f}M'
    else:
        return str(x)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(format_population))
plt.title('Top 5 Most Populous Countries', fontsize=16)
plt.xlabel('Country')
plt.ylabel('Population')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, format_population(yval, None),
             ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
genres = [
    'Action', 'Comedy', 'Drama', 'Horror', 'Romance',
    'Sci-Fi', 'Thriller', 'Animation', 'Adventure', 'Documentary'
]
movie_counts = [1200, 1500, 1800, 800, 1100, 950, 1000, 700, 900, 500]
plt.figure(figsize=(12, 6))
bars = plt.bar(genres, movie_counts, color='lightcoral')
plt.xticks(rotation=45, ha='right')
plt.title('Number of Movies by Genre', fontsize=16)
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 20, str(yval),
             ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```
